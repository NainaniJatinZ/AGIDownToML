# Normalizing Data
- more info on numpy.linalg.norm

```python

x = np.array([[0, 3, 4],
              [1, 6, 4]])
x_norm = np.linalg.norm(x, ord = 2, axis = 1, keepdims = True)
x = x/x_norm
print(x)
```
